```python
import pandas as pd

df = pd.read_pickle('df_numeric_encode.pkl')
df.head()
```

```python
# reverse the one-hot-encoding
# COMBINED_COVER, BUILDINGS_COVER, CONTENTS_COVER to a new feature COVER_TYPE
```

```python
df['COVER_TYPE'] = df.loc[:, ['COMBINED_COVER', 'BUILDINGS_COVER', 'CONTENTS_COVER']].idxmax(axis = 1)

df.head()
```

```python
df.drop(['COMBINED_COVER', 'BUILDINGS_COVER', 'CONTENTS_COVER'], axis = 1, inplace = True)
df.head()
```

```python
df.info()
```

```python
df_cover_cat = df['COVER_TYPE']
df_cover_cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df['COVER_TYPE_NUM'] = le.fit_transform(df_cover_cat)
df.drop('COVER_TYPE', axis = 1, inplace = True)

df.head()
```

```python
df.to_pickle('df_num_encode_2.pkl')
```

```python
y = df['RENEWED']
X = df.drop('RENEWED', axis = 1)
```

```python
X.head()
```

```python
# split into the training and testing datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
print(X_train.shape)
print(X_test.shape)

print(len(y_train))
print(len(y_test))
```

```python
X_train.to_pickle('X_train.pkl')
X_test.to_pickle('X_test.pkl')

y_train.to_pickle('y_train.pkl')
y_test.to_pickle('y_test.pkl')
```
